```
from urllib.request import urlretrieve
url = "https://github.com/Elwing-Chou/ml0716/raw/master/train.csv"
urlretrieve(url, "train.csv")
url = "https://github.com/Elwing-Chou/ml0716/raw/master/test.csv"
urlretrieve(url, "test.csv")
```

```
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
```

```
datas = pd.concat([train_df, test_df])
datas = datas.drop(["PassengerId", "Survived"], axis=1)
datas
```

```
# Step1. 補缺失值
s = datas.isna().sum()
# 篩選 series[[True....]]
s[s > 0]
```

```
most = datas["Embarked"].value_counts().idxmax()
datas["Embarked"] = datas["Embarked"].fillna(most)
```

```
# 先準備一下我要保留的list
def name(n):
    n = n.split(",")[-1].split(".")[0]
    return n.strip()
v = datas["Name"].apply(name).value_counts()
white = v.index[:4]
```

```
def name(n):
    n = n.split(",")[-1].split(".")[0]
    n = n.strip()
    if n in white:
        return n
    else:
        return None
datas["Name"] = datas["Name"].apply(name)
datas
```

```
v = datas["Ticket"].value_counts()
def ticket(t):
    global v
    return v[t]
datas["Ticket"] = datas["Ticket"].apply(ticket)
```

```
import numpy as np
def cabin(c):
    if pd.isna(c):
        return None
    else:
        return str(c)[0]
datas["Cabin"] = datas["Cabin"].apply(cabin)
```

```
# 如果你想要重製index: datas.reset_index(drop=True)
datas["Family"] = datas["Parch"] + datas["SibSp"]
datas
```

```
datas = pd.get_dummies(datas)
datas = pd.get_dummies(datas, columns=["Pclass"])
datas
```

```
datas = datas.fillna(datas.median())
```

```
datas
```

```
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# 取列: iloc(看實際是第幾列), loc(看列標籤)
# [第一列, 第二列....]
x_train = datas.iloc[:len(train_df)]
x_predict = datas.iloc[len(train_df):]
y_train = train_df["Survived"]
scaler = StandardScaler()
datas_scale = pd.DataFrame(scaler.fit_transform(datas), columns=datas.columns)
x_train_scale = datas_scale.iloc[:len(train_df)]
x_test_scale = datas_scale.iloc[len(train_df):]
```

```
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import numpy as np
```

```
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
params = {
    "n_neighbors":range(5, 50)
}
grid = GridSearchCV(clf, params, cv=10, n_jobs=4)
grid.fit(x_train_scale, y_train)
best2 = grid.best_estimator_
print("最佳參數:", grid.best_params_)
print("最佳分數:", grid.best_score_)
```

```
clf = KNeighborsClassifier(n_neighbors=11)
clf.fit(x_train_scale, y_train)
pre = clf.predict(x_test_scale)
pid = test_df["PassengerId"]
result = pd.DataFrame({
    "PassengerId":pid,
    "Survived":pre
})
result.to_csv("result_knn.csv", encoding="utf-8", index=False)
result
```
